```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from IPython.display import display, HTML

ruta = "/content/Telco_Customer_Churn_Muestra67 (1).xlsx"
df = pd.read_excel(ruta)

df_edad_1 = df[df["SeniorCitizen"] == 1]

display(HTML("<h1>An치lisis del Comportamiento de Clientes en una Empresa de Telecomunicaciones</h1>"))

display(HTML("""<h2>Dada una base de datos sobre una empresa de telecomunicaciones en Chile,
se utilizar치 Python para cargar y analizar una muestra representativa de estos datos.
Luego, a partir de la informaci칩n obtenida, se responder치n preguntas sobre mediciones estad칤sticas
con el fin de ayudar a la empresa a tomar decisiones informadas para dise침ar estrategias m치s efectivas
de fidelizaci칩n y segmentaci칩n de mercado.</h2>"""))

display(HTML("<h3>\n쮺u치l es la proporci칩n de clientes seniors y c칩mo se compara el gasto mensual promedio entre clientes seniors y no seniors?</h3>"))



print("\nEl numero de clientes (seniors/no seniors) es: ")

# N칰mero de seniors y no seniors totales
seniors = df[df["SeniorCitizen"] == 1]
no_seniors = df[df["SeniorCitizen"] == 0]

num_seniors = len(seniors)
num_no_seniors = len(no_seniors)

print(f"-Seniors: {num_seniors}")
print(f"-No seniors: {num_no_seniors}")

print("\nProporcion en porcentaje de cada tipo: ")

# Proporci칩n como porcentaje
total = num_seniors + num_no_seniors

proporcion_seniors = (num_seniors / total) * 100
proporcion_no_seniors = (num_no_seniors / total) * 100

print(f"-Seniors: {proporcion_seniors:.2f}%")
print(f"-No seniors: {proporcion_no_seniors:.2f}%")


print("\nEl promedio de gasto mensual por cada uno es: ")

# Promedio de gasto mensual
promedio_seniors = seniors["MonthlyCharges"].mean()
promedio_no_seniors = no_seniors["MonthlyCharges"].mean()

print(f"-Seniors: {promedio_seniors:.2f}")
print(f"-No seniors: {promedio_no_seniors:.2f}")

print("\nse ocupara un grafico de barras para comparar el gasto mensual promedio")

print("\n")
labels = ['Seniors', 'No Seniors']
promedios = [promedio_seniors, promedio_no_seniors]
plt.bar(labels, promedios, color=['salmon', 'skyblue'])
plt.title("Gasto mensual promedio por tipo de cliente")
plt.ylabel("Gasto mensual ($)")
plt.show()

display(HTML("""<h4>Con estos datos se puede ver que hay m치s clientes no seniors por un amplio porcentaje.
Respecto al gasto mensual de cada tipo, se observa que los no seniors son quienes m치s gastan cada mes.
Por ello, se podr칤a hacer m치s 칠nfasis en mantener y fidelizar a este tipo de clientes,
ya que representan una parte importante y rentable para la empresa.</h4>"""))


print("\n-----------------------------------------------------------------------")
display(HTML("""<h3>쯈u칠 porcentaje de clientes tiene contratado servicio de internet y c칩mo influye este factor en
su gasto mensual promedio?</h3>"""))
print("-----------------------------------------------------------------------")

print("El porcentaje de clientes con Internet contratado es:")

# Filtrar clientes con y sin Internet
con_internet = df[(df["InternetService"] == "DSL") | (df["InternetService"] == "Fiber optic")]
sin_internet = df[df["InternetService"] == "No"]

# Totales y porcentajes
total_internet = len(con_internet) + len(sin_internet)
porcentaje_con_internet = (len(con_internet) / total_internet) * 100
porcentaje_sin_internet = (len(sin_internet) / total_internet) * 100

print(f"- Con Internet: {porcentaje_con_internet:.2f}%")
print(f"- Sin Internet: {porcentaje_sin_internet:.2f}%")

# Gasto mensual promedio
promedio_con_internet = con_internet["MonthlyCharges"].mean()
promedio_sin_internet = sin_internet["MonthlyCharges"].mean()

print("\nGasto mensual promedio para cada segmento:")
print(f"- Con Internet: ${promedio_con_internet:.2f}")
print(f"- Sin Internet: ${promedio_sin_internet:.2f}")

# --- Gr치fico de barras simple ---
etiquetas = ['Con Internet', 'Sin Internet']
valores = [promedio_con_internet, promedio_sin_internet]

print("\n")
plt.figure(figsize=(6,5))
plt.bar(etiquetas, valores, color=['steelblue', 'salmon'])
plt.title("Gasto mensual promedio por tipo de cliente", fontsize=13)
plt.ylabel("Gasto mensual ($)", fontsize=11)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()

display(HTML("""<h4>Como se muestra en los datos y gr치ficos, las personas que contratan Internet
muy frecuentemente pagan m치s por mes. Con esto se puede inferir que es el servicio que m치s beneficios trae a la empresa,
por lo que es de mucha importancia darle visibilidad a este servicio.</h4>"""))



print("\n-----------------------------------------------------------------------")
print("Tercera pregunta")
print("-----------------------------------------------------------------------")

percentil25= df["tenure"].quantile(0.25)
percentil50 = df["tenure"].quantile(0.50)
percentil75 = df["tenure"].quantile(0.75)

print(f"Percentil 25: {percentil25}")
print(f"Percentil 50: {percentil50}")
print(f"Percentil 75: {percentil75}")



print("\n-----------------------------------------------------------------------")
print("Cuarta pregunta")
print("-----------------------------------------------------------------------")

cada_mes = df[df["Contract"] == "Month-to-month"]
un_a침o= df[df["Contract"] == "One year"]
dos_a침os = df[df["Contract"] == "Two year"]

total_cada_mes = len(cada_mes)
total_un_a침o = len(un_a침o)
total__dos_a침os = len(dos_a침os)

print(f"cada mes: {total_cada_mes}")
print(f"un a침o: {total_un_a침o}")
print(f"dos a침os: {total__dos_a침os}")

c_mas_frecuente = df["Contract"].mode()[0]

print(f"el contrato m치s frecuente es: {c_mas_frecuente}" + "(cada mes)")

print("\n-----------------------------------------------------------------------")
print("quinta pregunta")
print("-----------------------------------------------------------------------")

print("Para saber si hay correlacion, ocuparemos un grafico de dispersion, donde x es el gasto mensual y Y el total")


df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
df = df.dropna(subset=["TotalCharges"])

corr = df["MonthlyCharges"].corr(df["TotalCharges"])
print(f"Coeficiente de correlaci칩n: {corr:.2f}")


plt.scatter(df["MonthlyCharges"], df["TotalCharges"], alpha=0.5)
plt.title("Relaci칩n entre cargos mensuales y total cobrado")
plt.xlabel("Cargos mensuales ($)")
plt.ylabel("Total cobrado ($)")


x = df["MonthlyCharges"]
y = df["TotalCharges"]
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b, color="red")

plt.show()

print("\n-----------------------------------------------------------------------")
print("Sexta pregunta")
print("-----------------------------------------------------------------------")


contrato_1anio = df[df["Contract"] == "One year"]
contrato_2anios = df[df["Contract"] == "Two year"]


t_1anio = len(contrato_1anio)
t_2anios = len(contrato_2anios)

print(f"Clientes con contrato de un a침o: {t_1anio}")
print(f"Clientes con contrato de dos a침os: {t_2anios}")

print("\nVariabilidad del gasto mensual:")

print("\nContrato de 1 a침o")
print(contrato_1anio["MonthlyCharges"].describe())

print("\nContrato de 2 a침os")
print(contrato_2anios["MonthlyCharges"].describe())




print("\n-----------------------------------------------------------------------")
print("Septima pregunta")
print("-----------------------------------------------------------------------")

print("para saber si un cliente gasta mas o menos respecto a su tiempo de permanencia, ocuparemos un grafico de dispersion, donde x es el tiempo de permanencia y Y el gasto mensual")

plt.scatter(df["tenure"], df["MonthlyCharges"], alpha=0.5, color="steelblue", edgecolors="black")
plt.title("Relaci칩n entre tiempo de permanencia y gasto mensual")
plt.xlabel("Tiempo de permanencia (meses)")
plt.ylabel("Gasto mensual ($)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()


print("\n-----------------------------------------------------------------------")
print("octava pregunta")
print("-----------------------------------------------------------------------")

v = df[(df["InternetService"] == "DSL") | (df["InternetService"] == "Fiber optic")]
print(v["MonthlyCharges"].var())

print("\n-----------------------------------------------------------------------")
print("novena pregunta")
print("-----------------------------------------------------------------------")

plt.figure(figsize=(8,5))

for contrato, grupo in df.groupby("Contract"):
    plt.scatter(grupo["tenure"], grupo["MonthlyCharges"], label=contrato, alpha=0.6)

plt.title("Relaci칩n entre tiempo de permanencia y gasto mensual seg칰n tipo de contrato")
plt.xlabel("Tiempo de permanencia (meses)")
plt.ylabel("Gasto mensual ($)")
plt.legend(title="Tipo de contrato")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()

print("\n-----------------------------------------------------------------------")
print("decima pregunta")
print("-----------------------------------------------------------------------")

con_soporte = df[df["TechSupport"] == "Yes"]["MonthlyCharges"]
sin_soporte = df[df["TechSupport"] == "No"]["MonthlyCharges"]

# Calcular estad칤sticas
prom_con = con_soporte.mean()
prom_sin = sin_soporte.mean()
med_con = con_soporte.median()
med_sin = sin_soporte.median()
asim_con = con_soporte.skew()
asim_sin = sin_soporte.skew()

print("Comparaci칩n del gasto mensual seg칰n soporte t칠cnico\n")
print(f"俱뫮잺 Con soporte t칠cnico:")
print(f"Promedio: {prom_con:.2f}")
print(f"Mediana: {med_con:.2f}")
print(f"Asimetr칤a: {asim_con:.2f}\n")

print(f"俱뫮잺 Sin soporte t칠cnico:")
print(f"Promedio: {prom_sin:.2f}")
print(f"Mediana: {med_sin:.2f}")
print(f"Asimetr칤a: {asim_sin:.2f}")


print("\n-----------------------------------------------------------------------")
print("onceaba pregunta")
print("-----------------------------------------------------------------------")

pago_comun = df["PaymentMethod"].mode()[0]

print(f"el metodo de pago mas comun es: {pago_comun}")

print("como se relaciona el metodo de pago con el gasto mensual?")

metodo_comun = df["PaymentMethod"].mode()[0]
print(f"游눱 El m칠todo de pago m치s com칰n es: {metodo_comun}")

# Boxplot del gasto mensual seg칰n m칠todo de pago
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x="PaymentMethod", y="MonthlyCharges")
plt.title("Distribuci칩n del gasto mensual seg칰n m칠todo de pago")
plt.xlabel("M칠todo de pago")
plt.ylabel("Gasto mensual ($)")
plt.xticks(rotation=20)
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()











```
