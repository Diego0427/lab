```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

ruta = "/content/Telco_Customer_Churn_Muestra67 (1).xlsx"
df = pd.read_excel(ruta)

df_edad_1 = df[df["SeniorCitizen"] == 1]

print("¿Cuál es la proporción de clientes seniors y cómo se compara el gasto mensual promedio entre clientes seniors y no seniors?")

# Número de seniors y no seniors totales
seniors = df[df["SeniorCitizen"] == 1]
no_seniors = df[df["SeniorCitizen"] == 0]

num_seniors = len(seniors)
num_no_seniors = len(no_seniors)

print(f"número de seniors: {num_seniors}")
print(f"número de no seniors: {num_no_seniors}")

# Promedio de gasto mensual
promedio_seniors = seniors["MonthlyCharges"].mean()
promedio_no_seniors = no_seniors["MonthlyCharges"].mean()

print(f"promedio de seniors: {promedio_seniors:.2f}")
print(f"promedio de no seniors: {promedio_no_seniors:.2f}")

# Proporción como porcentaje
total = num_seniors + num_no_seniors

proporcion_seniors = (num_seniors / total) * 100
proporcion_no_seniors = (num_no_seniors / total) * 100

print(f"proporción de seniors: {proporcion_seniors:.2f}%")
print(f"proporción de no seniors: {proporcion_no_seniors:.2f}%")



print("\n-----------------------------------------------------------------------")
print("Segunda pregunta")
print("-----------------------------------------------------------------------")

print("el porcentaje de clientes con internet contratado es:")

con_internet = df[(df["InternetService"] == "DSL") | (df["InternetService"] == "Fiber optic")]
sin_internet = df[df["InternetService"] == "No"]

total_internet = len(con_internet) + len(sin_internet)

porcentaje_con_internet = (len(con_internet) / total_internet) * 100

print(f"porcentaje con internet: {porcentaje_con_internet:.2f}%")

print("\n-----------------------------------------------------------------------")
print("Tercera pregunta")
print("-----------------------------------------------------------------------")

percentil25= df["tenure"].quantile(0.25)
percentil50 = df["tenure"].quantile(0.50)
percentil75 = df["tenure"].quantile(0.75)

print(f"Percentil 25: {percentil25}")
print(f"Percentil 50: {percentil50}")
print(f"Percentil 75: {percentil75}")



print("\n-----------------------------------------------------------------------")
print("Cuarta pregunta")
print("-----------------------------------------------------------------------")

cada_mes = df[df["Contract"] == "Month-to-month"]
un_año= df[df["Contract"] == "One year"]
dos_años = df[df["Contract"] == "Two year"]

total_cada_mes = len(cada_mes)
total_un_año = len(un_año)
total__dos_años = len(dos_años)

print(f"cada mes: {total_cada_mes}")
print(f"un año: {total_un_año}")
print(f"dos años: {total__dos_años}")

c_mas_frecuente = df["Contract"].mode()[0]

print(f"el contrato más frecuente es: {c_mas_frecuente}" + "(cada mes)")

print("\n-----------------------------------------------------------------------")
print("quinta pregunta")
print("-----------------------------------------------------------------------")

print("Para saber si hay correlacion, ocuparemos un grafico de dispersion, donde x es el gasto mensual y Y el total")


df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
df = df.dropna(subset=["TotalCharges"])

corr = df["MonthlyCharges"].corr(df["TotalCharges"])
print(f"Coeficiente de correlación: {corr:.2f}")


plt.scatter(df["MonthlyCharges"], df["TotalCharges"], alpha=0.5)
plt.title("Relación entre cargos mensuales y total cobrado")
plt.xlabel("Cargos mensuales ($)")
plt.ylabel("Total cobrado ($)")


x = df["MonthlyCharges"]
y = df["TotalCharges"]
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b, color="red")

plt.show()

print("\n-----------------------------------------------------------------------")
print("Sexta pregunta")
print("-----------------------------------------------------------------------")


contrato_1anio = df[df["Contract"] == "One year"]
contrato_2anios = df[df["Contract"] == "Two year"]


t_1anio = len(contrato_1anio)
t_2anios = len(contrato_2anios)

print(f"Clientes con contrato de un año: {t_1anio}")
print(f"Clientes con contrato de dos años: {t_2anios}")

print("\nVariabilidad del gasto mensual:")

print("\nContrato de 1 año")
print(contrato_1anio["MonthlyCharges"].describe())

print("\nContrato de 2 años")
print(contrato_2anios["MonthlyCharges"].describe())




print("\n-----------------------------------------------------------------------")
print("Septima pregunta")
print("-----------------------------------------------------------------------")

print("para saber si un cliente gasta mas o menos respecto a su tiempo de permanencia, ocuparemos un grafico de dispersion, donde x es el tiempo de permanencia y Y el gasto mensual")

plt.scatter(df["tenure"], df["MonthlyCharges"], alpha=0.5, color="steelblue", edgecolors="black")
plt.title("Relación entre tiempo de permanencia y gasto mensual")
plt.xlabel("Tiempo de permanencia (meses)")
plt.ylabel("Gasto mensual ($)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()







```
